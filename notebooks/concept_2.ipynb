# Concept 2: Abstraction and Decomposition of Program Structure

## Task 1: Group lines into meaningful blocks

Below is a small Python program.

Your task is to divide the lines into **3 logical blocks**:
1. setup
2. processing
3. output

Write your answer as three groups of line numbers separated by semicolons.

Example format:

`1; 2,3; 4,5`

### Code

1. `numbers = [1, 2, 3, 4]`
2. `total = 0`
3. `for n in numbers:`
4. `    total = total + n`
5. `print(total)`


```python
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output

t1_correct_blocks = [['1', '2'], ['3', '4'], ['5']]

t1_answer_input = widgets.Text(
    value='',
    placeholder='e.g. 1,2; 3,4; 5',
    description='Blocks:',
    layout=widgets.Layout(width='420px')
)

t1_check_button = widgets.Button(description='Check answer', button_style='success')
t1_hint_button = widgets.Button(description='Show hint')
t1_solution_button = widgets.Button(description='Show solution')
t1_walkthrough_button = widgets.Button(description='Show walkthrough')

t1_output = widgets.Output()

def t1_normalize_blocks(text):
    groups = text.split(';')
    normalized = []
    for group in groups:
        items = [part.strip() for part in group.split(',') if part.strip()]
        if items:
            normalized.append(items)
    return normalized

def t1_on_check_clicked(_):
    with t1_output:
        clear_output()
        user_blocks = t1_normalize_blocks(t1_answer_input.value)

        if len(user_blocks) != 3:
            print('Please enter exactly 3 groups separated by semicolons.')
            return

        flat = [item for group in user_blocks for item in group]
        if sorted(flat) != ['1', '2', '3', '4', '5']:
            print('Please use each line number exactly once.')
            return

        if user_blocks == t1_correct_blocks:
            display(Markdown('**Correct.** The program can be decomposed into setup, processing, and output.'))
        else:
            display(Markdown('**Not quite.** Try thinking about which lines prepare state, which lines do the main work, and which line only presents the result.'))

def t1_on_hint_clicked(_):
    with t1_output:
        clear_output()
        display(Markdown('**Hint:** Look for lines that prepare values first, then lines that work over those values, and finally a line that only shows the result.'))

def t1_on_solution_clicked(_):
    with t1_output:
        clear_output()
        display(Markdown('**Solution:** `1,2; 3,4; 5`'))

def t1_on_walkthrough_clicked(_):
    with t1_output:
        clear_output()
        display(Markdown('## Walkthrough'))
        display(Markdown('**Block 1: Setup** → `numbers = [1, 2, 3, 4]` and `total = 0` create the data and initialize the accumulator.'))
        display(Markdown('**Block 2: Processing** → the `for` loop and its body perform the main computation by summing the numbers.'))
        display(Markdown('**Block 3: Output** → `print(total)` only displays the final result.'))

t1_check_button.on_click(t1_on_check_clicked)
t1_hint_button.on_click(t1_on_hint_clicked)
t1_solution_button.on_click(t1_on_solution_clicked)
t1_walkthrough_button.on_click(t1_on_walkthrough_clicked)

t1_controls = widgets.HBox([t1_check_button, t1_hint_button, t1_solution_button, t1_walkthrough_button])

display(t1_answer_input)
display(t1_controls)
display(t1_output)
```

## Task 2: Label the blocks by purpose

Below are three code blocks from a small program.

Your task is to match each block to its purpose.

Use these labels:
- setup
- processing
- output

Write your answer as:

`A=setup, B=processing, C=output`

### Block A
`scores = [7, 10, 4]`

### Block B
`total = sum(scores)`  
`average = total / len(scores)`

### Block C
`print(average)`

```python
t2_correct_answer = {'A': 'setup', 'B': 'processing', 'C': 'output'}

t2_answer_input = widgets.Text(
    value='',
    placeholder='e.g. A=setup, B=processing, C=output',
    description='Labels:',
    layout=widgets.Layout(width='500px')
)

t2_check_button = widgets.Button(description='Check answer', button_style='success')
t2_hint_button = widgets.Button(description='Show hint')
t2_solution_button = widgets.Button(description='Show solution')
t2_walkthrough_button = widgets.Button(description='Show walkthrough')

t2_output = widgets.Output()

def t2_parse_answer(text):
    result = {}
    parts = text.split(',')
    for part in parts:
        if '=' in part:
            left, right = part.split('=', 1)
            result[left.strip()] = right.strip().lower()
    return result

def t2_on_check_clicked(_):
    with t2_output:
        clear_output()
        user_answer = t2_parse_answer(t2_answer_input.value)

        if set(user_answer.keys()) != {'A', 'B', 'C'}:
            print('Please provide labels for A, B, and C.')
            return

        if user_answer == t2_correct_answer:
            display(Markdown('**Correct.** You identified the role of each block in the larger program structure.'))
        else:
            display(Markdown('**Not quite.** Think about which block creates data, which block transforms it, and which block only presents a result.'))

def t2_on_hint_clicked(_):
    with t2_output:
        clear_output()
        display(Markdown('**Hint:** Ask what each block is doing in the larger program. Is it creating data, transforming data, or only displaying something already computed?'))

def t2_on_solution_clicked(_):
    with t2_output:
        clear_output()
        display(Markdown('**Solution:** `A=setup, B=processing, C=output`'))

def t2_on_walkthrough_clicked(_):
    with t2_output:
        clear_output()
        display(Markdown('## Walkthrough'))
        display(Markdown('**Block A: Setup** → this block creates the initial data the rest of the program will use.'))
        display(Markdown('**Block B: Processing** → this block computes new values from the existing data.'))
        display(Markdown('**Block C: Output** → this block only displays the final result.'))

t2_check_button.on_click(t2_on_check_clicked)
t2_hint_button.on_click(t2_on_hint_clicked)
t2_solution_button.on_click(t2_on_solution_clicked)
t2_walkthrough_button.on_click(t2_on_walkthrough_clicked)

t2_controls = widgets.HBox([t2_check_button, t2_hint_button, t2_solution_button, t2_walkthrough_button])

display(t2_answer_input)
display(t2_controls)
display(t2_output)
```

## Task 3: Identify repeated logic and suggest an abstraction

Below is a small Python program.

Your task has two parts:

1. Identify which lines contain repeated logic
2. Suggest a short function name that could abstract that repeated work

Write your answer in this format:

`Repeated lines: 1,2 and 3,4`  
`Function name: show_doubled_value`

### Code

1. `double_a = a * 2`
2. `print(double_a)`
3. `double_b = b * 2`
4. `print(double_b)`

```python
t3_lines_input = widgets.Text(
    value='',
    placeholder='e.g. 1,2 and 3,4',
    description='Repeated:',
    layout=widgets.Layout(width='420px')
)

t3_name_input = widgets.Text(
    value='',
    placeholder='e.g. show_doubled_value',
    description='Function:',
    layout=widgets.Layout(width='420px')
)

t3_check_button = widgets.Button(description='Check answer', button_style='success')
t3_hint_button = widgets.Button(description='Show hint')
t3_solution_button = widgets.Button(description='Show solution')
t3_walkthrough_button = widgets.Button(description='Show walkthrough')

t3_output = widgets.Output()

def t3_on_check_clicked(_):
    with t3_output:
        clear_output()
        repeated_text = t3_lines_input.value.strip().lower()
        function_name = t3_name_input.value.strip()

        if not repeated_text or not function_name:
            print('Please answer both parts.')
            return

        normalized = repeated_text.replace(' ', '')
        acceptable_patterns = {'1,2and3,4', '3,4and1,2', '1,2&3,4', '3,4&1,2'}
        line_check = normalized in acceptable_patterns

        if line_check:
            display(Markdown(f'**Reasonable answer.** You identified repeated structure and suggested the abstraction `{function_name}`.'))
        else:
            display(Markdown('**Partly correct.** Look again at which lines repeat the same kind of work for different variables.'))

def t3_on_hint_clicked(_):
    with t3_output:
        clear_output()
        display(Markdown('**Hint:** Look for two chunks that do the same kind of work, but on different values. If the structure repeats, it may be a candidate for a helper function.'))

def t3_on_solution_clicked(_):
    with t3_output:
        clear_output()
        display(Markdown('**Possible solution:**'))
        display(Markdown('`Repeated lines: 1,2 and 3,4`'))
        display(Markdown('`Function name: show_doubled_value`'))
        display(Markdown('Other reasonable function names could also work.'))

def t3_on_walkthrough_clicked(_):
    with t3_output:
        clear_output()
        display(Markdown('## Walkthrough'))
        display(Markdown('Lines `1,2` and `3,4` have the same overall structure:'))
        display(Markdown('- compute a doubled value'))
        display(Markdown('- print that value'))
        display(Markdown('The only difference is whether the code works with `a` or `b`.'))
        display(Markdown('That repeated structure suggests that the work could be moved into a reusable function, such as `show_doubled_value`.'))

t3_check_button.on_click(t3_on_check_clicked)
t3_hint_button.on_click(t3_on_hint_clicked)
t3_solution_button.on_click(t3_on_solution_clicked)
t3_walkthrough_button.on_click(t3_on_walkthrough_clicked)

t3_controls = widgets.HBox([t3_check_button, t3_hint_button, t3_solution_button, t3_walkthrough_button])

display(t3_lines_input)
display(t3_name_input)
display(t3_controls)
display(t3_output)
```
