# Sequential Processes - Parsons Problem Prototype

You will be given a number of lines of Python code shown in the **wrong order**.

Your task is to decide the correct execution order.

Write your answer as a comma-separated sequence of line numbers, for example:

`2,1,3,4`

### Task 1: Variable Dependency
1. `z = y + 1`
2. `x = 2`
3. `y = x * 3`
4. `t = z * 2`

```python
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output

# Correct order:
# 2 -> x = 2
# 3 -> y = x * 3
# 1 -> z = y + 1
# 4 -> t = z * 2

vd_correct_order = ['2', '3', '1', '4']

vd_answer_input = widgets.Text(
    value='',
    placeholder='e.g. 3,2,4,1',
    description='Order:',
    layout=widgets.Layout(width='320px')
)

vd_check_button = widgets.Button(description='Check answer', button_style='success')
vd_hint_button = widgets.Button(description='Show hint')
vd_solution_button = widgets.Button(description='Show solution')
vd_walkthrough_button = widgets.Button(description='Show execution walkthrough')

vd_output = widgets.Output()

def vd_normalize_answer(text):
    return [part.strip() for part in text.split(',') if part.strip()]

def vd_on_check_clicked(_):
    with vd_output:
        clear_output()
        vd_user_answer = vd_normalize_answer(vd_answer_input.value)

        if len(vd_user_answer) != 4:
            print('Please enter exactly 4 line numbers, separated by commas.')
            return

        if sorted(vd_user_answer) != ['1', '2', '3', '4']:
            print('Please use each line number exactly once.')
            return

        if vd_user_answer == vd_correct_order:
            display(Markdown('**Correct.** The sequence produces a valid step-by-step execution flow.'))
        else:
            display(Markdown('**Incorrect.**'))

def vd_on_hint_clicked(_):
    with vd_output:
        clear_output()
        display(Markdown(
            '**Hint:** Start by finding the line that does not depend on any earlier variables. '
            'Then ask which variable must exist before each later line can run.'
        ))

def vd_on_solution_clicked(_):
    with vd_output:
        clear_output()
        display(Markdown('**Solution:** `2, 3, 1, 4`'))

def vd_on_walkthrough_clicked(_):
    with vd_output:
        clear_output()
        display(Markdown('## Execution walkthrough'))
        display(Markdown('1. `x = 2` → now `x` exists and has the value 2'))
        display(Markdown('2. `y = x * 3` → `y` becomes 6'))
        display(Markdown('3. `z = y + 1` → `z` becomes 7'))
        display(Markdown('4. `t = z * 2` → `t` becomes 14'))

vd_check_button.on_click(vd_on_check_clicked)
vd_hint_button.on_click(vd_on_hint_clicked)
vd_solution_button.on_click(vd_on_solution_clicked)
vd_walkthrough_button.on_click(vd_on_walkthrough_clicked)

vd_controls = widgets.HBox([vd_check_button, vd_hint_button, vd_solution_button, vd_walkthrough_button])

display(vd_answer_input)
display(vd_controls)
display(vd_output)
```

### Task 2: Conditional Logic
1. `if age >= 18:`
2. `message = "adult"`
3. `age = 20`
4. `status = message`

```python
# Correct order:
# 3 -> age = 20
# 1 -> if age >= 18:
# 2 -> message = "adult"
# 4 -> status = message

cl_correct_order = ['3', '1', '2', '4']

cl_answer_input = widgets.Text(
    value='',
    placeholder='e.g. 3,2,4,1',
    description='Order:',
    layout=widgets.Layout(width='320px')
)

cl_check_button = widgets.Button(description='Check answer', button_style='success')
cl_hint_button = widgets.Button(description='Show hint')
cl_solution_button = widgets.Button(description='Show solution')
cl_walkthrough_button = widgets.Button(description='Show execution walkthrough')

cl_output = widgets.Output()

def cl_normalize_answer(text):
    return [part.strip() for part in text.split(',') if part.strip()]

def cl_on_check_clicked(_):
    with cl_output:
        clear_output()
        cl_user_answer = cl_normalize_answer(cl_answer_input.value)

        if len(cl_user_answer) != 4:
            print('Please enter exactly 4 line numbers, separated by commas.')
            return

        if sorted(cl_user_answer) != ['1', '2', '3', '4']:
            print('Please use each line number exactly once.')
            return

        if cl_user_answer == cl_correct_order:
            display(Markdown('**Correct.** The sequence produces a valid step-by-step execution flow.'))
        else:
            display(Markdown('**Incorrect.**'))

def cl_on_hint_clicked(_):
    with cl_output:
        clear_output()
        display(Markdown(
            '**Hint:** Look for the value that the condition depends on. '
            'That value must exist before the condition can be checked.'
        ))

def cl_on_solution_clicked(_):
    with cl_output:
        clear_output()
        display(Markdown('**Solution:** `3, 1, 2, 4`'))

def cl_on_walkthrough_clicked(_):
    with cl_output:
        clear_output()
        display(Markdown('## Execution walkthrough'))
        display(Markdown('1. `age = 20` → now `age` exists and has the value 20'))
        display(Markdown('2. `if age >= 18:` → the condition is checked and evaluates to `True`'))
        display(Markdown('3. `message = "adult"` → because the condition was true, `message` is created with the value `"adult"`'))
        display(Markdown('4. `status = message` → `status` now gets the value `"adult"`'))

cl_check_button.on_click(cl_on_check_clicked)
cl_hint_button.on_click(cl_on_hint_clicked)
cl_solution_button.on_click(cl_on_solution_clicked)
cl_walkthrough_button.on_click(cl_on_walkthrough_clicked)

cl_controls = widgets.HBox([cl_check_button, cl_hint_button, cl_solution_button, cl_walkthrough_button])

display(cl_answer_input)
display(cl_controls)
display(cl_output)
```

### Task 3: Loop Setup
1. `total = total + n`
2. `numbers = [1, 2, 3]`
3. `total = numbers[0]`
4. `for n in numbers[1:]:`

```python
# Correct order:
# 2 -> numbers = [1, 2, 3]
# 3 -> total = numbers[0]
# 4 -> for n in numbers[1:]:
# 1 -> total = total + n

ls_correct_order = ['2', '3', '4', '1']

ls_answer_input = widgets.Text(
    value='',
    placeholder='e.g. 3,2,4,1',
    description='Order:',
    layout=widgets.Layout(width='320px')
)

ls_check_button = widgets.Button(description='Check answer', button_style='success')
ls_hint_button = widgets.Button(description='Show hint')
ls_solution_button = widgets.Button(description='Show solution')
ls_walkthrough_button = widgets.Button(description='Show execution walkthrough')

ls_output = widgets.Output()

def ls_normalize_answer(text):
    return [part.strip() for part in text.split(',') if part.strip()]

def ls_on_check_clicked(_):
    with ls_output:
        clear_output()
        ls_user_answer = ls_normalize_answer(ls_answer_input.value)

        if len(ls_user_answer) != 4:
            print('Please enter exactly 4 line numbers, separated by commas.')
            return

        if sorted(ls_user_answer) != ['1', '2', '3', '4']:
            print('Please use each line number exactly once.')
            return

        if ls_user_answer == ls_correct_order:
            display(Markdown('**Correct.** The sequence produces a valid step-by-step execution flow.'))
        else:
            display(Markdown('**Incorrect.**'))

def ls_on_hint_clicked(_):
    with ls_output:
        clear_output()
        display(Markdown(
            '**Hint:** Finding the line that creates the sequence. '
            'Ask which value must exist before the loop can update it.'
        ))

def ls_on_solution_clicked(_):
    with ls_output:
        clear_output()
        display(Markdown('**Solution:** `2, 3, 4, 1`'))

def ls_on_walkthrough_clicked(_):
    with ls_output:
        clear_output()
        display(Markdown('## Execution walkthrough'))
        display(Markdown('1. `numbers = [1, 2, 3]` → now the list `numbers` exists'))
        display(Markdown('2. `total = numbers[0]` → `total` starts with the first value, so `total = 1`'))
        display(Markdown('3. `for n in numbers[1:]:` → the loop runs over the remaining values: `2` and `3`'))
        display(Markdown('4. `total = total + n` → inside the loop, `total` is updated each time'))
        display(Markdown('   - when `n = 2`, `total` becomes 3'))
        display(Markdown('   - when `n = 3`, `total` becomes 6'))

ls_check_button.on_click(ls_on_check_clicked)
ls_hint_button.on_click(ls_on_hint_clicked)
ls_solution_button.on_click(ls_on_solution_clicked)
ls_walkthrough_button.on_click(ls_on_walkthrough_clicked)

ls_controls = widgets.HBox([ls_check_button, ls_hint_button, ls_solution_button, ls_walkthrough_button])

display(ls_answer_input)
display(ls_controls)
display(ls_output)
```
